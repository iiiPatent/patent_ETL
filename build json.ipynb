```python
#ID
#step 2
def merge_keywords(id_keywords_file_path, group_dic):
    word_list=[]
    with open(id_keywords_file_path, 'r') as f:
        all_patent= f.read()
    f.close()
    for one_patent in all_patent.split('\n'):
        one_patent_id=one_patent.split(",")[0].strip()
        one_patent_keyword=one_patent.split(",")[1].strip()
        for one_word in one_patent_keyword.split(" "):
            word_list.append(one_word)
        new_value =list(group_dic[one_patent_id]) + word_list
        
        group_dic[one_patent_id]=new_value   
    print "step 2 finish"
    return group_dic
```

```python
#160  7
#step 3
def source_to_big_dic(source_dic):
    big_dic={}
    count=0
    for _id in source_dic:  #_id
        count+=1
        print count
        I_got_this = 0  #
    #============================================================================
        for group_num in big_dic:    #
            if source_dic[_id][0]==group_num:    #id
                I_got_this = 1               #

    #===================================================================================
                big_dic[source_dic[_id][0]][source_dic[_id][0]+"_id"].append(_id)
                for one_words in source_dic[_id]:                #_ID
                    if one_words not in big_dic[group_num]:   #   
                        big_dic[group_num][one_words] = 1      #
                    else: 
                        big_dic[group_num][one_words]=big_dic[group_num][one_words]+1  #

    #========================================="0~n"_idKEY========================     
        if I_got_this == 0:         #
            big_dic[source_dic[_id][0]]={}     #big_dickey ()
            big_dic[source_dic[_id][0]][source_dic[_id][0]+"_id"]=[]    #"0_id"list
            group_num=source_dic[_id][0]       #

    #==================================================================================
            big_dic[source_dic[_id][0]][source_dic[_id][0]+"_id"].append(_id)  
            for one_words in source_dic[_id]:                #_ID
                if one_words not in big_dic[group_num]:   #   
                    big_dic[group_num][one_words] = 1      #
                else: 
                    big_dic[group_num][one_words]=big_dic[group_num][one_words]+1  #
    print "step 3 finish"
    return big_dic
```
